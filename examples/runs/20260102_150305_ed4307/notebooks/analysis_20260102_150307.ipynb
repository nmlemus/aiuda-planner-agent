## Setup & Imports


````python
from datetime import datetime, timedelta
import warnings

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
````

## Analysis


### Design dataset structure (brands, products, time periods, metrics)


````python
warnings.filterwarnings('ignore')

# Set random seed for reproducibility
np.random.seed(42)

# Define parameters
brands = ['Brand A', 'Brand B', 'Brand C', 'Brand D', 'Brand E']
product_categories = ['Electronics', 'Home Appliances', 'Personal Care', 'Kitchen']
regions = ['North', 'South', 'East', 'West', 'Central']
months = pd.date_range(start='2022-01-01', end='2023-12-31', freq='MS')

# Create comprehensive dataset
data_records = []

for month in months:
    month_num = month.month
    
    for brand in brands:
        for category in product_categories:
            for region in regions:
                # Base sales with brand-specific market positions
                brand_strength = {
                    'Brand A': 1.0,   # Our brand - moderate position
                    'Brand B': 1.2,   # Market leader
                    'Brand C': 0.9,   # Close competitor
                    'Brand D': 0.7,   # Smaller player
                    'Brand E': 0.6    # Smallest player
                }
                
                # Category preferences by brand
                category_strength = {
                    'Brand A': {'Electronics': 1.1, 'Home Appliances': 0.9, 'Personal Care': 1.0, 'Kitchen': 0.95},
                    'Brand B': {'Electronics': 1.2, 'Home Appliances': 1.1, 'Personal Care': 0.9, 'Kitchen': 1.0},
                    'Brand C': {'Electronics': 0.9, 'Home Appliances': 1.0, 'Personal Care': 1.2, 'Kitchen': 0.95},
                    'Brand D': {'Electronics': 0.8, 'Home Appliances': 0.85, 'Personal Care': 0.9, 'Kitchen': 1.1},
                    'Brand E': {'Electronics': 0.7, 'Home Appliances': 0.8, 'Personal Care': 0.85, 'Kitchen': 0.9}
                }
                
                # Seasonal patterns (higher in Q4, lower in Q2)
                seasonal_factor = 1.0 + 0.3 * np.sin((month_num - 3) * np.pi / 6)
                
                # Growth trend (Brand A growing faster than others)
                months_since_start = (month - months[0]).days / 30
                if brand == 'Brand A':
                    growth_factor = 1.0 + 0.015 * months_since_start  # 1.5% monthly growth
                elif brand == 'Brand B':
                    growth_factor = 1.0 + 0.005 * months_since_start  # Slower growth
                else:
                    growth_factor = 1.0 + 0.008 * months_since_start
                
                # Base sales volume
                base_volume = 1000 * brand_strength[brand] * category_strength[brand][category]
                sales_volume = base_volume * seasonal_factor * growth_factor * np.random.uniform(0.85, 1.15)
                
                # Pricing strategy
                base_prices = {
                    'Electronics': 500,
                    'Home Appliances': 350,
                    'Personal Care': 50,
                    'Kitchen': 150
                }
                
                brand_price_premium = {
                    'Brand A': 1.0,
                    'Brand B': 1.15,  # Premium pricing
                    'Brand C': 0.95,
                    'Brand D': 0.85,
                    'Brand E': 0.80
                }
                
                avg_price = base_prices[category] * brand_price_premium[brand] * np.random.uniform(0.95, 1.05)
                revenue = sales_volume * avg_price
                
                # Cost structure (Brand A has improving margins)
                cost_ratio = 0.65 if brand == 'Brand A' else np.random.uniform(0.60, 0.70)
                cost = revenue * cost_ratio
                profit = revenue - cost
                
                # Customer satisfaction (Brand A improving)
                base_satisfaction = {
                    'Brand A': 7.5 + 0.05 * months_since_start,
                    'Brand B': 8.2,
                    'Brand C': 7.8,
                    'Brand D': 7.0,
                    'Brand E': 6.8
                }
                customer_satisfaction = min(10, base_satisfaction[brand] + np.random.uniform(-0.5, 0.5))
                
                # Market share calculation will be done later
                data_records.append({
                    'date': month,
                    'brand': brand,
                    'product_category': category,
                    'region': region,
                    'sales_volume': int(sales_volume),
                    'revenue': round(revenue, 2),
                    'avg_price': round(avg_price, 2),
                    'cost': round(cost, 2),
                    'profit': round(profit, 2),
                    'customer_satisfaction': round(customer_satisfaction, 1),
                    'year': month.year,
                    'quarter': f'Q{(month.month-1)//3 + 1}',
                    'month_name': month.strftime('%B')
                })

# Create DataFrame
df = pd.DataFrame(data_records)

# Calculate market share by month and category
df['market_share'] = 0.0
for date in df['date'].unique():
    for category in df['product_category'].unique():
        mask = (df['date'] == date) & (df['product_category'] == category)
        total_revenue = df.loc[mask, 'revenue'].sum()
        df.loc[mask, 'market_share'] = (df.loc[mask, 'revenue'] / total_revenue * 100).round(2)

# Save dataset
df.to_csv('data/brand_sales_dataset.csv', index=False)

print("✓ Synthetic dataset created successfully!")
print(f"\nDataset shape: {df.shape}")
print(f"Date range: {df['date'].min()} to {df['date'].max()}")
print(f"\nBrands: {df['brand'].unique().tolist()}")
print(f"Product categories: {df['product_category'].unique().tolist()}")
print(f"Regions: {df['region'].unique().tolist()}")
print(f"\nFirst few rows:")
print(df.head(10))
print(f"\nDataset statistics:")
print(df.describe())
````

````python
# Exploratory Data Analysis
print("="*80)
print("EXPLORATORY DATA ANALYSIS - BRAND A FOCUS")
print("="*80)

# Overall market position
print("\n1. OVERALL MARKET POSITION (Total Revenue)")
print("-" * 60)
brand_revenue = df.groupby('brand')['revenue'].sum().sort_values(ascending=False)
brand_revenue_pct = (brand_revenue / brand_revenue.sum() * 100).round(2)
for brand, rev in brand_revenue.items():
    pct = brand_revenue_pct[brand]
    print(f"{brand:12s}: ${rev:,.0f} ({pct}%)")

# Brand A specific metrics
print("\n2. BRAND A KEY METRICS")
print("-" * 60)
brand_a_data = df[df['brand'] == 'Brand A']
print(f"Total Revenue: ${brand_a_data['revenue'].sum():,.0f}")
print(f"Total Profit: ${brand_a_data['profit'].sum():,.0f}")
print(f"Profit Margin: {(brand_a_data['profit'].sum() / brand_a_data['revenue'].sum() * 100):.2f}%")
print(f"Avg Customer Satisfaction: {brand_a_data['customer_satisfaction'].mean():.2f}/10")
print(f"Total Sales Volume: {brand_a_data['sales_volume'].sum():,} units")

# Growth analysis
print("\n3. YEAR-OVER-YEAR GROWTH (2022 vs 2023)")
print("-" * 60)
for brand in brands:
    brand_data = df[df['brand'] == brand]
    rev_2022 = brand_data[brand_data['year'] == 2022]['revenue'].sum()
    rev_2023 = brand_data[brand_data['year'] == 2023]['revenue'].sum()
    growth = ((rev_2023 - rev_2022) / rev_2022 * 100)
    print(f"{brand:12s}: {growth:+.2f}%")

# Category performance for Brand A
print("\n4. BRAND A PERFORMANCE BY CATEGORY")
print("-" * 60)
brand_a_category = brand_a_data.groupby('product_category').agg({
    'revenue': 'sum',
    'profit': 'sum',
    'market_share': 'mean',
    'customer_satisfaction': 'mean'
}).round(2)
print(brand_a_category)

# Regional performance for Brand A
print("\n5. BRAND A PERFORMANCE BY REGION")
print("-" * 60)
brand_a_region = brand_a_data.groupby('region').agg({
    'revenue': 'sum',
    'profit': 'sum',
    'sales_volume': 'sum'
}).round(2)
brand_a_region['profit_margin_%'] = (brand_a_region['profit'] / brand_a_region['revenue'] * 100).round(2)
print(brand_a_region)

# Competitive positioning
print("\n6. AVERAGE MARKET SHARE BY BRAND")
print("-" * 60)
avg_market_share = df.groupby('brand')['market_share'].mean().sort_values(ascending=False)
for brand, share in avg_market_share.items():
    print(f"{brand:12s}: {share:.2f}%")

print("\n✓ EDA completed successfully!")
````

### Create market share analysis visualizations


````python
# Set style for professional visualizations
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")

# Create market share analysis visualizations
fig, axes = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle('Market Share Analysis - Brand A Competitive Position', fontsize=18, fontweight='bold', y=0.995)

# 1. Overall Market Share Pie Chart
ax1 = axes[0, 0]
market_share_total = df.groupby('brand')['revenue'].sum()
colors = ['#2ecc71' if brand == 'Brand A' else '#3498db' if brand == 'Brand B' else '#95a5a6' 
          for brand in market_share_total.index]
explode = [0.1 if brand == 'Brand A' else 0 for brand in market_share_total.index]

wedges, texts, autotexts = ax1.pie(market_share_total, labels=market_share_total.index, autopct='%1.1f%%',
                                     colors=colors, explode=explode, startangle=90, textprops={'fontsize': 11})
for autotext in autotexts:
    autotext.set_color('white')
    autotext.set_fontweight('bold')
ax1.set_title('Overall Market Share by Revenue\n(2022-2023)', fontsize=13, fontweight='bold', pad=15)

# 2. Market Share Trend Over Time
ax2 = axes[0, 1]
monthly_share = df.groupby(['date', 'brand'])['revenue'].sum().reset_index()
monthly_share['total_revenue'] = monthly_share.groupby('date')['revenue'].transform('sum')
monthly_share['market_share_pct'] = (monthly_share['revenue'] / monthly_share['total_revenue'] * 100)

for brand in brands:
    brand_data = monthly_share[monthly_share['brand'] == brand]
    linewidth = 3 if brand == 'Brand A' else 1.5
    alpha = 1.0 if brand == 'Brand A' else 0.6
    linestyle = '-' if brand == 'Brand A' else '--'
    ax2.plot(brand_data['date'], brand_data['market_share_pct'], 
             label=brand, linewidth=linewidth, alpha=alpha, linestyle=linestyle,
             marker='o' if brand == 'Brand A' else None, markersize=4)

ax2.set_xlabel('Date', fontsize=11, fontweight='bold')
ax2.set_ylabel('Market Share (%)', fontsize=11, fontweight='bold')
ax2.set_title('Market Share Trend Over Time', fontsize=13, fontweight='bold', pad=15)
ax2.legend(loc='best', framealpha=0.9)
ax2.grid(True, alpha=0.3)
ax2.tick_params(axis='x', rotation=45)

# 3. Market Share by Product Category
ax3 = axes[1, 0]
category_share = df.groupby(['product_category', 'brand'])['revenue'].sum().reset_index()
category_total = category_share.groupby('product_category')['revenue'].sum().reset_index()
category_total.columns = ['product_category', 'total_revenue']
category_share = category_share.merge(category_total, on='product_category')
category_share['market_share_pct'] = (category_share['revenue'] / category_share['total_revenue'] * 100)

category_pivot = category_share.pivot(index='product_category', columns='brand', values='market_share_pct')
category_pivot = category_pivot[brands]  # Ensure consistent order

x = np.arange(len(category_pivot.index))
width = 0.15

for i, brand in enumerate(brands):
    color = '#2ecc71' if brand == 'Brand A' else None
    ax3.bar(x + i*width, category_pivot[brand], width, label=brand, color=color, alpha=0.8)

ax3.set_xlabel('Product Category', fontsize=11, fontweight='bold')
ax3.set_ylabel('Market Share (%)', fontsize=11, fontweight='bold')
ax3.set_title('Market Share by Product Category', fontsize=13, fontweight='bold', pad=15)
ax3.set_xticks(x + width * 2)
ax3.set_xticklabels(category_pivot.index, rotation=15, ha='right')
ax3.legend(loc='best', framealpha=0.9, fontsize=9)
ax3.grid(True, alpha=0.3, axis='y')

# 4. Market Share Growth Rate
ax4 = axes[1, 1]
# Calculate quarterly market share
df['year_quarter'] = df['year'].astype(str) + '-' + df['quarter']
quarterly_share = df.groupby(['year_quarter', 'brand'])['revenue'].sum().reset_index()
quarterly_share['total_revenue'] = quarterly_share.groupby('year_quarter')['revenue'].transform('sum')
quarterly_share['market_share_pct'] = (quarterly_share['revenue'] / quarterly_share['total_revenue'] * 100)

# Calculate growth rate
growth_rates = []
for brand in brands:
    brand_quarterly = quarterly_share[quarterly_share['brand'] == brand].sort_values('year_quarter')
    first_quarter = brand_quarterly.iloc[0]['market_share_pct']
    last_quarter = brand_quarterly.iloc[-1]['market_share_pct']
    growth_rate = ((last_quarter - first_quarter) / first_quarter * 100)
    growth_rates.append(growth_rate)

colors_growth = ['#2ecc71' if brand == 'Brand A' else '#e74c3c' if rate < 0 else '#3498db' 
                 for brand, rate in zip(brands, growth_rates)]
bars = ax4.barh(brands, growth_rates, color=colors_growth, alpha=0.8)
ax4.axvline(x=0, color='black', linestyle='-', linewidth=0.8)
ax4.set_xlabel('Market Share Growth Rate (%)', fontsize=11, fontweight='bold')
ax4.set_title('Market Share Growth Rate (Q1 2022 to Q4 2023)', fontsize=13, fontweight='bold', pad=15)
ax4.grid(True, alpha=0.3, axis='x')

# Add value labels
for i, (bar, rate) in enumerate(zip(bars, growth_rates)):
    ax4.text(rate + (1 if rate > 0 else -1), i, f'{rate:+.1f}%', 
             va='center', ha='left' if rate > 0 else 'right', fontweight='bold')

plt.tight_layout()
plt.savefig('artifacts/01_market_share_analysis.png', dpi=300, bbox_inches='tight')
plt.show()

print("✓ Market share analysis visualizations saved to artifacts/01_market_share_analysis.png")
````

````python
# Create profitability and pricing analysis visualizations
fig, axes = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle('Profitability & Pricing Analysis - Brand A Financial Performance', fontsize=18, fontweight='bold', y=0.995)

# 1. Profit Margin Comparison
ax1 = axes[0, 0]
brand_financials = df.groupby('brand').agg({
    'revenue': 'sum',
    'profit': 'sum',
    'cost': 'sum'
}).reset_index()
brand_financials['profit_margin'] = (brand_financials['profit'] / brand_financials['revenue'] * 100)

colors_margin = ['#2ecc71' if brand == 'Brand A' else '#3498db' for brand in brand_financials['brand']]
bars = ax1.bar(brand_financials['brand'], brand_financials['profit_margin'], color=colors_margin, alpha=0.8)
ax1.set_xlabel('Brand', fontsize=11, fontweight='bold')
ax1.set_ylabel('Profit Margin (%)', fontsize=11, fontweight='bold')
ax1.set_title('Profit Margin Comparison by Brand', fontsize=13, fontweight='bold', pad=15)
ax1.grid(True, alpha=0.3, axis='y')
ax1.tick_params(axis='x', rotation=45)

# Add value labels
for bar in bars:
    height = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2., height,
             f'{height:.1f}%', ha='center', va='bottom', fontweight='bold')

# 2. Revenue vs Profit Scatter
ax2 = axes[0, 1]
for brand in brands:
    brand_data = df[df['brand'] == brand].groupby('date').agg({
        'revenue': 'sum',
        'profit': 'sum'
    }).reset_index()
    
    size = 100 if brand == 'Brand A' else 50
    alpha = 0.8 if brand == 'Brand A' else 0.4
    marker = 'o' if brand == 'Brand A' else 's'
    
    ax2.scatter(brand_data['revenue']/1e6, brand_data['profit']/1e6, 
                s=size, alpha=alpha, label=brand, marker=marker)

ax2.set_xlabel('Revenue ($ Millions)', fontsize=11, fontweight='bold')
ax2.set_ylabel('Profit ($ Millions)', fontsize=11, fontweight='bold')
ax2.set_title('Revenue vs Profit Relationship', fontsize=13, fontweight='bold', pad=15)
ax2.legend(framealpha=0.9, loc='best')
ax2.grid(True, alpha=0.3)

# Add trend line for Brand A
brand_a_monthly = df[df['brand'] == 'Brand A'].groupby('date').agg({
    'revenue': 'sum',
    'profit': 'sum'
}).reset_index()
z = np.polyfit(brand_a_monthly['revenue']/1e6, brand_a_monthly['profit']/1e6, 1)
p = np.poly1d(z)
x_line = np.linspace(brand_a_monthly['revenue'].min()/1e6, brand_a_monthly['revenue'].max()/1e6, 100)
ax2.plot(x_line, p(x_line), "g--", linewidth=2, alpha=0.6, label='Brand A Trend')

# 3. Price Positioning Matrix
ax3 = axes[1, 0]
brand_price_volume = df.groupby('brand').agg({
    'avg_price': 'mean',
    'sales_volume': 'sum',
    'revenue': 'sum'
}).reset_index()

for brand in brands:
    brand_row = brand_price_volume[brand_price_volume['brand'] == brand].iloc[0]
    size = brand_row['revenue'] / 1e6  # Size proportional to revenue
    color = '#2ecc71' if brand == brand else '#3498db'
    alpha = 0.9 if brand == 'Brand A' else 0.5
    
    ax3.scatter(brand_row['avg_price'], brand_row['sales_volume']/1000, 
                s=size*2, alpha=alpha, label=brand, color=color)
    
    # Add brand labels
    ax3.annotate(brand, 
                 (brand_row['avg_price'], brand_row['sales_volume']/1000),
                 xytext=(5, 5), textcoords='offset points', fontweight='bold', fontsize=9)

ax3.set_xlabel('Average Price ($)', fontsize=11, fontweight='bold')
ax3.set_ylabel('Total Sales Volume (Thousands)', fontsize=11, fontweight='bold')
ax3.set_title('Price-Volume Positioning (Bubble size = Revenue)', fontsize=13, fontweight='bold', pad=15)
ax3.grid(True, alpha=0.3)

# Add quadrant lines
avg_price_all = brand_price_volume['avg_price'].mean()
avg_volume_all = brand_price_volume['sales_volume'].mean() / 1000
ax3.axvline(x=avg_price_all, color='gray', linestyle='--', linewidth=1, alpha=0.5)
ax3.axhline(y=avg_volume_all, color='gray', linestyle='--', linewidth=1, alpha=0.5)

# 4. Profit Trend Over Time
ax4 = axes[1, 1]
monthly_profit = df.groupby(['date', 'brand'])['profit'].sum().reset_index()

for brand in brands:
    brand_data = monthly_profit[monthly_profit['brand'] == brand]
    linewidth = 3 if brand == 'Brand A' else 1.5
    alpha = 1.0 if brand == 'Brand A' else 0.5
    marker = 'o' if brand == 'Brand A' else None
    markersize = 5 if brand == 'Brand A' else 0
    ax4.plot(brand_data['date'], brand_data['profit']/1e6, 
             label=brand, linewidth=linewidth, alpha=alpha, marker=marker, markersize=markersize)

ax4.set_xlabel('Date', fontsize=11, fontweight='bold')
ax4.set_ylabel('Monthly Profit ($ Millions)', fontsize=11, fontweight='bold')
ax4.set_title('Monthly Profit Trend by Brand', fontsize=13, fontweight='bold', pad=15)
ax4.legend(loc='best', framealpha=0.9)
ax4.grid(True, alpha=0.3)
ax4.tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.savefig('artifacts/05_profitability_pricing.png', dpi=300, bbox_inches='tight')
plt.show()

print("✓ Profitability and pricing analysis saved to artifacts/05_profitability_pricing.png")
````
